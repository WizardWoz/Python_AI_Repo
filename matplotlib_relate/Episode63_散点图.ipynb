```python
# 散点图（X-Y图）：将所有的数据以点的形式展现在直角坐标系上，以显示变量之间的相互影响程度，点的位置由变量的数值决定
# 通过散点图可以直观地看出变量之间的关系，如线性关系、非线性关系、正相关、负相关等。
from matplotlib import pyplot as plt
import numpy as np
# 可通过修改字体配置来支持中文显示
plt.rcParams['font.sans-serif']=['simhei']      # 负号没了
# 当字体设置成中文后，必须设置特殊符号的编码，不然无法正常显示
plt.rcParams['axes.unicode_minus']=False        # 不使用Unicode的负号
```

```python
# 1.普通散点图&折线图
# x轴数据
x=np.array([1,2,3,4,5,6,7,8])
# y轴数据
y=np.array([1,4,9,16,7,11,23,18])
# 绘制散点图
plt.scatter(x,y)
plt.show()
plt.plot(x,y)  # 绘制折线图
```

```python
# 2.带颜色和透明度的散点图
# x轴数据
x=np.array([1,2,3,4,5,6,7,8])
# y轴数据
y=np.array([1,4,9,16,7,11,23,18])
# 生成一个[0,1)之间的随机浮点数或N维整数
s=(20*np.random.rand(8))**2
print(s)  # 打印每个点的大小
# 绘制散点图
# alpha透明度；color颜色
colors=np.random.rand(8)  # 生成8个随机颜色
plt.scatter(x,y,s=s,alpha=0.3,c=colors)
plt.show()
# plt.plot(x,y)  # 绘制折线图
```

```python
# 3.连续化色图
# x轴数据
x=np.random.rand(100)  # 生成1到100的整数
# y轴数据
y=np.random.rand(100)  # 生成1到100的整数
# 颜色随机
colors=np.arange(1,101)
plt.scatter(x,y,c=colors,cmap='viridis')
```
