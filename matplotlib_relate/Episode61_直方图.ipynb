```python
# 直方图，又称质量分布图，是条形图的一种，一般用于描述连续型数据的分布关系（柱状图是离散型）
# 由一系列高度不等的纵向线段表示数据分布的情况。横轴表示数据类型，纵轴表示分布情况
# 用于概率分布，显示了一组数值序列在给定数值范围内出现的概率；而柱状图则用于展示各个类别的频数
from matplotlib import pyplot as plt
import numpy as np
# 可通过修改字体配置来支持中文显示
plt.rcParams['font.sans-serif']=['simhei']      # 负号没了
# 当字体设置成中文后，必须设置特殊符号的编码，不然无法正常显示
plt.rcParams['axes.unicode_minus']=False        # 不使用Unicode的负号
```

```python
# 1.柱状直方图
# 使用numpy随机生成300个整数
x_value=np.random.randint(140,180,300)
print(x_value)
# 绘制直方图
num,bin,patches=plt.hist(x_value,bins=10,edgecolor='white')
plt.title('数据统计')
plt.xlabel('身高')
plt.ylabel('比率')
print(num)  # 每个区间的数量，共10个区间
print(bin)  # 每个区间的边界，共11个边界
print(patches)  # 每个区间的矩形对象，共10个矩形对象
for p in patches:
    print(p)
    print(p.get_x(),p.get_y(),p.get_width(),p.get_height())  # 矩形的左下角坐标和宽高
```

```python
# 2.折线直方图
# 创建一个画布
fig, ax = plt.subplots()
# 绘制直方图
num, bins_limit, patches = ax.hist(x_value, bins=10, edgecolor='white')
# 注意num返回的个数是10个，bin返回的边界是11个，需要截取
bins_limit = bins_limit[:-1]  # 截取前10个边界
print(bins_limit)
# 绘制折线图
ax.plot(bins_limit, num, '--',marker='o')
```

```python
# 3.不等距分组：只需确定分组上下限，并制定histtype='bar'
fig,ax = plt.subplots()
x=np.random.normal(100,20,100)
bins=[50,60,70,90,100,110,140,150]
ax.hist(x,bins,color='g',rwidth=0.4)
ax.set_title('不等距分组直方图')
plt.show()
```

```python
# 4.多类型直方图
n_bins = 10
fig, ax = plt.subplots(figsize=(8,5))
# 分别生成10000，5000,2000个值（使用生成器）
x_multi=[np.random.randn(n) for n in [10000, 5000, 2000]]
print(x_multi)
# 实际绘图代码与单类型直方图差异不大，只是增加了一个图例项
# 在ax.hist函数中指定图例label名称
ax.hist(x_multi,n_bins,histtype='bar',label=list("ABC"))
ax.set_title('多类型直方图')
ax.legend()  # 显示图例
```

```python
# 5.堆叠直方图
x_value=np.random.randint(140,180,200)
x2_value=np.random.randint(140,180,200)
# 默认stacked=False，不堆叠
plt.hist([x_value,x2_value], bins=10, stacked=True)
```
