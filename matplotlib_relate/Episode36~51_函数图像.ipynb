```python
# 数据可视化：将数据转换成图表的形式，更直观地展示数据；可准确高效、简洁全面地传递信息，帮助我们发现规律和特征，挖掘数据背后的价值
# 根据numpy.ndarray数组绘制2D/3D图像，使用简单，代码清晰易懂
# 柱状图=条形图；直方图=特殊的条形图；散点图；饼状图
# 优点：
# (1).matplotlib提供了一套面向绘图对象编程的API接口
# (2).借助Python强大的可扩展性，matplot可在许多不同的环境被调用
# (3).基于matlab和基于面向对象，但完全免费
# (4).matlotlib实现了几乎完全自主控制的图形定义功能
import matplotlib
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

```python
# 1.首先导入matplotlib包的pyplot模块，并以as 别名的形式简化引入包名
from matplotlib import pyplot as plt
# 使用numpy提供的arange()方法创建需要绘图的数据
x=np.arange(-10,11)     # 所得的值应用于x轴
# 在同一个坐标系中绘制多个图像
y1=x**2
y2=x
print(x)
print(y1)
print(y2)
```

```python
# 1.应先设置好图表相关参数，再进行绘图
# 为图表添加名称
plt.title("y=x^2")
# 标题设置为中文，默认不支持中文，显示：missing from current font.
# 可通过修改字体配置来支持中文显示
plt.rcParams['font.sans-serif']=['simhei']      # 负号没了
# 当字体设置成中文后，必须设置特殊符号的编码，不然无法正常显示
plt.rcParams['axes.unicode_minus']=False        # 不使用Unicode的负号
plt.title("y等于x的平方")
# 设置x轴、y轴名称；设置字体
plt.xlabel("x轴",fontsize=12)
plt.ylabel("y轴")

# 2.plot(横轴,纵轴)：绘制线性图标，并设置线条
plt.plot(x,y1,linewidth=5)
plt.plot(x,y2)
```

```python
# 每个时间点的销量绘图
times=['2015/6/26','2015/8/1','2015/9/6','2015/10/12','2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/17',
       '2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']
# 随机生成销量数据
sales=np.random.randint(500,2001,size=len(times))
# 绘制图形
# plt.plot(times,sales)

# 3.xticks()、yticks()：设置x轴、y轴的刻度
# 按照某个规则显示
plt.xticks(range(1,len(times),2),rotation=45)  # 从第1个日期开始到最后，每间隔一个日期打印一次
plt.plot(times,sales)
```

```python
# 每个时间点的销量绘图
times=['2015/6/26','2015/8/1','2015/9/6','2015/10/12','2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/17',
       '2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']
# 随机生成收入
income=np.random.randint(500,2001,size=len(times))
# 随机生成支出
expense=np.random.randint(300,1501,size=len(times))

# 绘制图形
plt.xticks(range(1,len(times)),rotation=45)
# 使用图例前为每个图形设置label参数
plt.plot(times,income,label='收入')
plt.plot(times,expense,label='支出')
# 6.plt.legend()：默认使用每个图像的label值作为图例说明，默认位置location=best，不会遮挡图象
plt.legend(loc='best')

# 7.plt.text()：显示每条数据的值以及x、y值的位置
# zip()：将两个list列表/ndarray数组压缩为一个元组tuple
for a,b in zip(times,income):
    plt.text(a,b,'%s万'%b)
for a,b in zip(times,expense):
    plt.text(a,b,'%s万'%b)
```

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
c,s=np.cos(x),np.sin(x)
plt.plot(x,c)
plt.plot(x,s)
# 8.plt.grid()：显示网格
# linestyle：线型（虚线/实线）；color：颜色；linewidth：线宽；axis：x轴方向/y轴方向/整体
plt.grid(linestyle='--')
```

```python
# 9.对坐标轴的操作
x=np.arange(-50,51)
y=x**2
# 获取当前坐标轴
ax=plt.gca()
# 通过坐标轴spines，确定top、bottom、left、right
# 不需要右侧和上侧线条，可以通过设置颜色为None使其消失
ax.spines['right'].set_color("none")
ax.spines['top'].set_color("none")
# position的位置参数有三种：data，outward（向外），axes（0.0~1.0），整个轴上的比例
ax.spines['left'].set_position(('data',0.0))
# ax.spines['left'].set_position(('outward',0.4))
# 设置坐标区间
plt.ylim(0,y.max())
plt.plot(x,y)
```
