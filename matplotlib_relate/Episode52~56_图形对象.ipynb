```python
# 创建图形对象：面向对象编程的核心思想；通过图形对象调用其它方法和属性，有助于我们更好地处理多个画布；
# pyplot负责生成图形对象，并通过该对象添加一个或多个axes对象（即绘图区域）
# 图形类模块matplotlib.figure：包含创建图形对象的方法figure()
from matplotlib import pyplot as plt    # 等价于import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1.创建图形对象
fig=plt.figure()
```

```python
# 通过plt.rcParams['figure.dpi']改变图形对象的分辨率
plt.rcParams['figure.dpi']=300
fig=plt.figure()
```

```python
fig=plt.figure(figsize=(4,2),facecolor='g')
# 2.创建区域，并在可绘图区域中作画
# axl从画布起始位置绘制，宽高和画布一致
ax1=fig.add_axes([0,0,1,1])
# 区域1作画
ax1.plot([1,2,3,4,6],[2,3,5,8,9])
ax1.plot([1,2,3,4,6],[9,8,7,6,5])
# 创建区域2，当前的默认区域是区域2，使用plt.plot()绘制的图像在此区域
# ax2从画布20%的位置开始绘制，宽高是画布的50%
ax2=fig.add_axes([0.2,0.2,0.5,0.5])
# 区域2作画
ax2.plot([1,2,3,4,6],[9,8,7,6,5])
# 不能直接在图形对象（画布）上作画，以下代码会报错，必须先创建区域
# fig.plot()
```

```python
fig=plt.figure()
# 3.得到当前的可绘图区域
ax=plt.gca()
# plot(y)，x可省略，默认[0...N-1]递增，N为y轴元素个数
plt.plot(range(50,71),marker='o')
plt.grid()
# y轴的长度为20
y_len=len(range(50,70))
print(np.linspace(0,y_len-1,y_len))
# subplot()方法：均等划分画布
ax.plot()
```

```python
# 4.创建一个子图，表示一个有2行1列的网格的顶部图
# 在创建时设置多图的基本信息
plt.subplot(211,title="pic1",xlabel='x axis')
# x可以省略，默认[0,1,...,N-1]递增
plt.plot(range(50,71))
# 创建另一个子图
plt.subplot(212,title="pic2",xlabel='x axis')
plt.plot(np.arange(12)**2)
# 发现子图标题重叠，最后使用plt.tight_layout()
plt.tight_layout()
```

```python
# 使用pyplot()模块中的方法设置好再绘制
# 第一个区域
plt.subplot(211)
plt.title("ax1")
plt.plot(range(50,71))
# 第二个区域
plt.subplot(212)
plt.title("ax2")
plt.plot(np.arange(12)**2)
plt.tight_layout()
```

```python
# 5.subplots()：与subplot()类似，既创建了一个包含子图区域的画布，又创建了一个figure对象；subplot()只创建一个包含子图区域的画布
# 创建2行2列的子图，返回图形对象（画布）、所有子图的坐标轴
fig,axes=plt.subplots(2,2)
# 第一个区域ax1
ax1=axes[0][0]
# x轴
x=np.arange(1,5)
# 绘制平方函数
ax1.plot(x,x*x)
ax1.set_title('square')

# 第二个区域：绘制平方根函数
axes[0][1].plot(x,np.sqrt(x))
axes[0][1].set_title('square root')

# 第三个区域：绘制指数函数
axes[1][0].plot(x,np.exp(x))
axes[1][0].set_title('exp')

# 第四个区域：绘制对数函数
axes[1][1].plot(x,np.log10(x))
axes[1][1].set_title('log')

# 处理标题遮挡
plt.tight_layout()
```
