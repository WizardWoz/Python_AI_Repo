```python
# 柱状图的绘制
from matplotlib import pyplot as plt
import numpy as np
# 可通过修改字体配置来支持中文显示
plt.rcParams['font.sans-serif']=['simhei']      # 负号没了
# 当字体设置成中文后，必须设置特殊符号的编码，不然无法正常显示
plt.rcParams['axes.unicode_minus']=False        # 不使用Unicode的负号
```

```python
# 1.绘制简单的柱状图
# x轴数据
x=range(5)
# y轴数据
data=[5,20,15,25,10]
# 设置图形标题
plt.title("基本柱状图")
# 绘制网格
plt.grid(ls="--",alpha=0.5)
# bar绘制图形，x表示坐标，data为表示柱状图的高度
plt.bar(x,data,bottom=[10,20,5,0,10])
# 注意data=[5,20,15,25,10]对应的bottom[10,20,5,0,10]
# 会在y轴原来的数据基础上加上bottom[]对应位置的数据
```

```python
# 设置颜色
# x轴数据
x=range(5)
# y轴数据
data=[5,20,15,25,10]
# 设置图形标题
plt.title("基本柱状图")
# 绘制网格
plt.grid(ls="--",alpha=0.5)
# bar绘制图形，x表示坐标，data为表示柱状图的高度
plt.bar(x,data,color=["r",'g','b'])
```

```python
# 2.plt.bar()：同位置多柱状图
# (1)设置好数据
# 国家
countries=['挪威','德国','中国','美国','瑞典']
# 金牌个数
gold_medal=[16,12,9,8,8]
# 银牌个数
silver_medal=[8,10,4,10,5]
# 铜牌个数
bronze_medal=[13,5,2,7,5]
# (2)将x轴转换为数值
x=np.arange(len(countries))
print(x)
# (3)每种奖牌的色条宽度
width=0.2
# (4)确定每种奖牌x轴坐标的起始位置
# 金牌起始位置
gold_x=x
# 银牌起始位置
silver_x=x+width
# 铜牌起始位置
bronze_x=x+2*width
# (5)绘制各种奖牌图形
# 金牌图形
plt.bar(gold_x,gold_medal,width=width,color='gold')
# 银牌图形
plt.bar(silver_x,silver_medal,width=width,color='silver')
# 铜牌图形
plt.bar(bronze_x,bronze_medal,width=width,color='saddlebrown')
# (6)优化x轴的坐标
plt.xticks(x+width,labels=countries)
# 显示奖牌数目
for i in range(len(countries)):
    # 金牌的数目，va即vertical alignment垂直对齐位置；ha即horizonal alignment水平对齐位置
    plt.text(gold_x[i],gold_medal[i],gold_medal[i],va="bottom",ha="center")
    plt.text(silver_x[i],silver_medal[i],silver_medal[i],va="bottom",ha="center")
    plt.text(bronze_x[i],bronze_medal[i],bronze_medal[i],va="bottom",ha="center")
```

```python
# 3.plt.barh()：简单水平条形柱状图
# (1)设置好数据
# 国家
movie=['挪威','德国','中国','美国','瑞典']
# (3)每种奖牌的色条宽度
width=0.2
# 金牌个数
gold_medal=[16,12,9,8,8]
plt.barh(countries,width=gold_medal)
```

```python
# 4.水平堆叠柱状图
movie=['新蝙蝠侠','狙击手','奇迹笨小孩']
# 第一天
real_day1=np.array([4053,2548,1543])
# 第二天
real_day2=np.array([7840,4013,2421])
# 第三天
real_day3=np.array([8080,3673,1342])

left_day2=real_day1     # 第二天距离左侧的为第一天的数值
left_day3=real_day1+real_day2   # 第三天距离左侧的为第一天+第二天的数值
# 设置线条高度
height=0.2
# 绘制图形
# 第一天图形
plt.barh(movie,real_day1,height=height)
# 第二天图形
plt.barh(movie,real_day2,left=left_day2,height=height)
# 第三天图形
plt.barh(movie,real_day3,left=left_day3,height=height)

# 设置数值文本：计算宽度值和y轴的值
sum_data=real_day1+real_day2+real_day3
# va即vertical alignment垂直对齐位置：控制文本的x位置参数表示文本边界框的左边、中间或右边
# ha即horizonal alignment水平对齐位置：控制文本的y位置参数表示文本边界框的底部、中间或顶部
for i in range(len(movie)):
    plt.text(sum_data[i],movie[i],sum_data[i],ha='left',va='center')
plt.xlim(0,sum_data.max()+2000)
```
