```python
# 箱线图：一种用作显示一组数据分布情况的图形，主要用于显示数据的中位数、四分位数、异常值等统计信息
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 1.绘制简单箱线图
# x=np.random.rand(20) # 生成20个随机数
x=np.array([1,20,30,50,60])
# showmeans：是否显示均值
# meanline（可选参数）：均值显示是线还是点，类型为布尔值，默认值为False；当取值为True且showmeans、shownotches都为True时，均值显示为线
# labels：每个数据集的标签
plt.boxplot(x,showmeans=True,meanline=True)  # 绘制箱线图
plt.grid()  # 添加网格线
plt.show()  # 显示图形
```

```python
# 2.为一维数据绘制箱线图
# 创建5行5列的数据
x=np.random.randint(10,100,(5,5))
box={"linestyle":"--","linewidth":1,"color":"blue"} # 设置箱线图的样式
mean={"marker":"o","markerfacecolor":"pink","markersize":2}  # 设置均值的样式
# boxprops：箱体的样式，类型为字典
# meanprops：算术平均值的样式，类型为字典
plt.boxplot(x,boxprops=box,meanprops=mean,showmeans=True,meanline=True,labels=['A','B','C','D','E'])  # 绘制箱线图
plt.grid()  # 添加网格线
plt.show()  # 显示图形
```

```python
# 3.为二维数据绘制箱线图
x=np.array([1,2,3,4,5,3,4,5,7,1,8,7]).reshape(3,4)  # 生成3行4列的数据
print(x)
plt.boxplot(x,showmeans=True,meanline=True,labels=list("ABCD"))  # 绘制箱线图
plt.grid()  # 添加网格线
plt.show()  # 显示图形
```
